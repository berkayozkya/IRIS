#nicel özellikler

```python
def Mean(u):
  mean = sum(u)/len(u)
  return mean
def Standard_Deviatoin(u):
  total = 0
  for i in u:
    total += ((Mean(u)-i) ** 2)
  return sqrt(total/(len(u)-1))
def Maximum(u):
  maximum=0
  for i in u:
    if i>maximum:
        maximum=i
  return maximum
def Minimum(u):
  minimum=Maximum(u)
  for i in u:
    if i<minimum:
        minimum=i
  return minimum
```

#kovaryans

```python
def Coveriance(u,v):
  total=0
  for i,j in zip(u,v):
    total+=(Mean(u)-i)*(Mean(v)-j)
  return total/(len(u)-1)
```

```python
data.cov()
```

```python
Coveriance(data['sepal_length'],data['sepal_length'])
```

#Corralation

```python
def Correlation(u,v): 
  return Coveriance(u,v)/(Standard_Deviatoin(u)*Standard_Deviatoin(v))
```
